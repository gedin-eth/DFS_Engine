## import modules

```python
import numpy as np
import pandas as pd
import time

timestr =time.strftime("%d%m%y")
```

```python
def clean1(self):
    self.rename({"Unnamed: 0":"a"}, axis="columns", inplace=True)
    self.drop(["a"], axis=1, inplace=True)
    
def clean2(self):
    self.rename({"Unnamed: 0.1":"a"}, axis="columns", inplace=True)
    self.drop(["a"], axis=1, inplace=True)
```

```python

def exposure(final):
    
    position = ['PG', 'SG', 'SF', 'PF', 'C', 'G', 'F', 'UTIL']
    player = set([])
    
    for p in position:
        player.update(set(final[p]))

    
    b =[]
    c =[]
    for i in player:
        count = 0
        for j in position:
            for k in final[j]:
                if i ==k:
                    count = count+1
        b.append(i)
        c.append((count/len(final))*100)

    report = pd.DataFrame()
    report.insert(0,'Player',b)
    report.insert(1,'Exposure', c)

    return report.sort_values(by= 'Exposure', ascending = False)
```
