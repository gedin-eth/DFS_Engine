```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import time

timestr =time.strftime("%d%m%y")
```

# import files

```python
# importing lineups

line = pd.read_csv('poop{date}.csv'.format(date=timestr))
# importing player pool
kay = pd.read_csv('output2{date}.csv'.format(date=timestr))
```

# simulation setup

```python
#creating an empty dataframe to store key features for simulation
kay_b = pd.DataFrame()
```

```python
# storing the NameIds from the player pool in a column named Name
#kay_b.insert(0,'Name', kay.nameid)
kay_b.insert(0,'Name', kay.nameid)
```

```python
# storing the Projections from the player pool in a column named proj
kay_b.insert(1,'proj', kay.proj)
```

```python
# storing the standard deviations from the player pool in a column named std
kay_b.insert(2, 'std', kay['std'])
```

```python
# storing the odds from the player pool in a column named odds
kay_b.insert(3,'odds', kay['odds'])
```

```python
# storing the ceilings from the player pool in a column named ceil
kay_b.insert(4,'ceil',kay['ceil'])
```

```python
kay_b.insert(5,'player', kay.name)
```

```python
def PosNormal(mean, sigma):
    x = np.random.normal(mean,sigma,1)
    return(x if x>=0 else PosNormal(mean,sigma))
```

## removing lineups with players not in my pool

```python
kay90 = kay_b.copy()

for i in range(len(kay)):
     kay90.proj[i] = PosNormal(kay_b.iloc[i,1] ,kay_b.iloc[i,2])
        #print(kay1.val[i])
    
    # The players nameid and new projections are bundled into a dictionary
    # so they can be mapped to the lineups dataframe. Once mapped we can caculate
    # each lineups totals
kay2 = dict(zip(kay90.Name, kay90.proj))
    

    ## lineup being mapped by position and data is saved in new columns
line['p1']=line.PG.map(kay2)
line['p2']=line.SG.map(kay2)
line['p3']=line.SF.map(kay2)
line['p4']=line.PF.map(kay2)
line['p5']=line.C.map(kay2)
line['p6']=line.G.map(kay2)
line['p7']=line.F.map(kay2)
line['p8']=line.UTIL.map(kay2)
    
    ## the sum from the the columns to calculate lineup total
line['tot'] = line.p1 + line.p2 + line.p3 + line.p4 + line.p5 + line.p6 + line.p7 + line.p8
```

```python
line = line[line['tot'].notna()]
```

```python
num_sim = 1000  # num_sim is the number of times the simulation will run. For typical monte carlo simulations
                # at least 1000 runs are needed
    
    
# num_l is the number of lineups     
num_l = len(line)   
```

## running simulation

```python
### need to speed up this step
### 1. try generating all of the values, then mapping
### 2. is there a more efficient pandas function for this

#can turn this into a callable function. Also Need to fix the slice on index issue

sim = pd.DataFrame() #creating an empty dataframe to store simulation results (Rankings)

val = pd.DataFrame() #creating an empty dataframe to store simulation results (Points)
kay1 = kay_b.copy()  #copying the key features into a new dataFrame


## This for loop will run until it reaches the number set as "num_sim"

for j in range(num_sim):

    ## this for loop will run untill every entry in 
    ##the player pool has new projected points based off their point distribution
    for i in range(len(kay)):
        kay1.proj[i] = PosNormal(kay_b.iloc[i,1] ,kay_b.iloc[i,2])
        #print(kay1.val[i])
    
    # The players nameid and new projections are bundled into a dictionary
    # so they can be mapped to the lineups dataframe. Once mapped we can caculate
    # each lineups totals
    kay2 = dict(zip(kay1.Name, kay1.proj))
    

    ## lineup being mapped by position and data is saved in new columns
    line['p1']=line.PG.map(kay2)
    line['p2']=line.SG.map(kay2)
    line['p3']=line.SF.map(kay2)
    line['p4']=line.PF.map(kay2)
    line['p5']=line.C.map(kay2)
    line['p6']=line.G.map(kay2)
    line['p7']=line.F.map(kay2)
    line['p8']=line.UTIL.map(kay2)
    
    ## the sum from the the columns to calculate lineup total
    line['tot'] = line.p1 + line.p2 + line.p3 + line.p4 + line.p5 + line.p6 + line.p7 + line.p8
    
    # saving results in the data frame
    sim.insert(j,j, line.tot.rank(ascending = 0))
    val.insert(j,j,line.tot)
    
    
    # user print out .... could request the elasped time and ETA
    if j == 1:
        
        print('Monte Carlo Simulation is Loading....')
        
    elif j==round(num_sim*.1):
        print('Monte Carlo Simulation is 10% complete')
        
    elif j==round(num_sim*.25):
        print('Monte Carlo Simulation is 25% complete')
    elif j== round(num_sim*.5):
        print('Monte Carlo Simulation is 50% complete')
    elif j== round(num_sim*.75):
        print('Monte Carlo Simulation is 75% complete')
    elif j== (num_sim-1):
        print('Monte Carlo Simulation is 100% complete')
```

## post processing results

```python
# simR may need to be deleted
#need to convert this into a function

simR = pd.DataFrame()

# l is a dataFrame that will hold the total points scored by each lineup
l = pd.DataFrame()


## a - m stores diffrent features of the simulation results

a = []
b = []
c = []
d = []
e = []
f = []
g = []
h = []
k = []
m = []

token = True  #GPP or cash
for i in range(len(sim)):
    if token == True:
        p = sim.iloc[i,:][sim.iloc[i,:]<= round(num_l*0.01)] #top 1% 12
        q = sim.iloc[i,:][sim.iloc[i,:]<=round(num_l*0.1)] #top 10% 117
        r = sim.iloc[i,:][sim.iloc[i,:]>round(num_l*0.1)] #remaining 117
        l = val.iloc[i,:]
        
        a.append(i+1)
        b.append(p.count())
        e.append((p.count()/num_sim)*100)
        c.append(q.count())
        f.append((q.count()/num_sim)*100)
        d.append(r.count())
        h.append((r.count()/num_sim)*100)
        g.append((p.count()*400)+(q.count()*80)-(r.count()*40))
        k.append(l.mean())
        m.append(l.std())
        
    else:
        p = sim.iloc[i,:][sim.iloc[i,:]<=600] #top 50%
        r = sim.iloc[i,:][sim.iloc[i,:]>600] #bottom 50%
        l = val.iloc[i,:]

        
        a.append(i+1)
        b.append(p.count())
        e.append((p.count()/num_sim)*100)
        d.append(r.count())
        h.append((r.count()/num_sim)*100)
        g.append((p.count()*5)-(r.count()*5))
        k.append(l.mean())
        m.append(l.std())
        
        
        
    if i == 1:
        
        print('Cleaning Data')
        
    elif i==round(num_l*.1):
        print('Data is 10% clean')
        
    elif i==round(num_l*.25):
        print('Data is 25% clean')
    elif i== round(num_l*.5):
        print('Data is 50% clean')
    elif i== round(num_l*.75):
        print('Data is 75% clean')
    elif i== (num_l-1):
        print('Data is 100% clean')
```
